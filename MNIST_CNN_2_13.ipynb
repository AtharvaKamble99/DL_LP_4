```python
import pandas as pd 
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras.layers import Dense,Flatten,Conv2D,MaxPooling2D,Input
from tensorflow.keras.utils import to_categorical
from tensorflow.keras import models,backend
```

```python
train_data=pd.read_csv("mnist_train.csv")
test_data=pd.read_csv("mnist_test.csv")
```

```python
train_data
```

```python
test_data
```

```python
xtrain=train_data.drop(['label'],axis=1).values
xtest=test_data.drop(['label'],axis=1).values
```

```python
ytrain= train_data['label'].values
ytest = test_data['label'].values
```

```python
xtrain = xtrain.reshape(-1,28,28,1).astype('float32')/255.0
```

```python
xtest= xtest.reshape(-1,28,28,1).astype('float32')/255.0
```

```python
ytrain= to_categorical(ytrain,10)
```

```python
ytest= to_categorical(ytest,10)
```

```python
xtrain.shape
```

```python
model=models.Sequential(
    [
        Input((28,28,1)),
        Conv2D(32,(3,3),activation='relu'),
        MaxPooling2D((2,2)),
        Conv2D(64,(3,3),activation='relu'),
        MaxPooling2D((2,2)),
        Conv2D(64,(3,3),activation='relu'),
        Flatten(),
        Dense(64,activation='relu'),
        Dense(10,activation='softmax')
    ]
)
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
r=model.fit(xtrain,ytrain,epochs=5,validation_data=(xtest,ytest))
```

```python
loss,acc= model.evaluate(xtest,ytest)
```

```python
loss
```

```python
acc
```

```python
from sklearn.metrics import accuracy_score
predict=model.predict(xtest)
```

```python
ypred_label=np.argmax(predict,axis=1)
ytest_label= np.argmax(ytest,axis=1)
```

```python
print(accuracy_score(ytest_label,ypred_label))
```
